# Data Cleaning

```python
df_clean = df.copy()
```

```python
df_clean.isnull().sum()
```

```python
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())
```

```python
df_clean['Age'].isnull().sum()
```

```python
df_clean['City'] = df_clean['City'].fillna(df_clean['City'].mode()[0])
```

```python
df_clean['City'].isnull().sum()
```

```python
df_clean['Order_Date'] = pd.to_datetime(df_clean['Order_Date'])
```

```python
df_clean.info()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.isnull().sum()
```

# Feature Engineering

```python
df_clean['Order_Month'] = df_clean['Order_Date'].dt.month_name()
```

```python
df_clean[['Order_Date', 'Order_Month']].head()
```

```python
df_clean['Order_Year'] = df_clean['Order_Date'].dt.year
```

```python
df_clean[['Order_Date', 'Order_Year']].head()
```

```python
df_clean["Order_Quarter"] = "Q" + df_clean["Order_Date"].dt.quarter.astype(str)
```

```python
df_clean[["Order_Date", "Order_Quarter"]].head()
```
